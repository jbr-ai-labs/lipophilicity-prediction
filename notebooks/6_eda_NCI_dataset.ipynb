## Summary

1. There are 98% NA values for experimantal logP

2. There are 2534 unique molecules with experimental logP value

3. There is no Temperature and pH columns

```python
from rdkit.Chem import PandasTools
import pandas as pd
import os
from rdkit import RDConfig
```

```python
DATASET_PATH = '../data/raw'
DATASET_FILENAME = 'NCIdb.sdf'
```

```python
sdfFile = os.path.join(DATASET_PATH ,DATASET_FILENAME)
frame = PandasTools.LoadSDF(sdfFile,smilesName='SMILES',molColName='Molecule',
           includeFingerprints=False)
```

```python
frame.info()
```

```python
frame = frame.drop(columns = list(set(list(frame))-set(['SMILES','Experimental logP'])))
```

```python
frame.info()
```

```python
frame['logP'] = pd.to_numeric(frame['Experimental logP'])
frame['smiles'] = frame['SMILES']

PROCESSED_PATH = '../data/1_filtering/'

frame[['smiles','logP']].to_csv(os.path.join(PROCESSED_PATH,'NCIDataset.csv'))
```

## Getting info about dataset

```python
import pandas as pd
import os
import seaborn as sns
```

```python
DATASET_PATH = '../data/1_filtering/'
DATASET_FILENAME = 'NCIDataset.csv'
```

```python
logp_data = pd.read_csv(os.path.join(DATASET_PATH,DATASET_FILENAME), index_col = 0)
```

```python
VALUE_COLUMN = "logP"

SMILES_COLUMN = "smiles"

n = logp_data.shape[0]
```

```python
# check NA values
num_of_na_value = logp_data[VALUE_COLUMN].isna().sum()
frac_of_na_value = int(100.0*(num_of_na_value / n))
print("# of NA values logP = " + str(num_of_na_value) + ", " + str(frac_of_na_value) + "%")
logp_data[logp_data[VALUE_COLUMN].isna()]
```

```python
# drop NA values
logp_data = logp_data[logp_data[VALUE_COLUMN].notna()]
```
